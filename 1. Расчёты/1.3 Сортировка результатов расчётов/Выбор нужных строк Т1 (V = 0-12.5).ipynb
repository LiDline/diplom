```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import numpy as np
import plotly.io as pio # Useng for animation
from collections import Counter
from scipy import interpolate
```

```python
file = 'C:/Users/Burger/source/work/double-propellers-and-rings/2СВК/T/Calculation Results/T1/Vн = 0-12.5/25.csv'
new_df = pd.read_csv(file, sep = ',',skiprows=3)
new_df =new_df.apply(pd.to_numeric, errors = 'coerce')

df = pd.DataFrame(new_df.iloc[:,1:])
df.rename({'P3': 'Vн [км/ч]', 'P1': 'h1 [град]', 'P4': 'n1 [об/мин]', 'P2': 'h2 [град]', 'P5': 'n2 [об/мин]',
            'P6': 'ТВ1 [Н]','P7': 'ТВ2 [Н]', 'P8': 'RК1 [Н]','P9': 'RК2 [Н]',
             'P10': 'R ВМГ [Н]', 'P11': 'R тела [Н]','P12': 'R суммарная [Н]', 
             'P13': 'Mx1 [Н*м]', 'P14': 'Mx2 [Н*м]',  'P15': 'Mx ВМГ [Н*м]', 'P16': 'Mx суммарная [Н*м]',
             }, axis=1, inplace=True) 
df.loc[(df['n1 [об/мин]'] == 0.01), 'n1 [об/мин]'] = 0 
df.loc[(df['n2 [об/мин]'] == -0.01), 'n2 [об/мин]'] = 0    

df = df.reindex(columns=['Vн [км/ч]', 'h1 [град]', 'n1 [об/мин]', 'h2 [град]', 'n2 [об/мин]', 'ТВ1 [Н]','ТВ2 [Н]', 'RК1 [Н]','RК2 [Н]',
             'R ВМГ [Н]', 'R тела [Н]','R суммарная [Н]', 
            'Mx1 [Н*м]', 'Mx2 [Н*м]',  'Mx ВМГ [Н*м]', 'Mx суммарная [Н*м]',])

df.head()
```

```python
df_done = pd.DataFrame()
Vx = 0



for i in range (0,6):               # Смотрю куски по Vн
    df_Vx = pd.DataFrame()
    df_Vx = df[(df['Vн [км/ч]'] == Vx)]
    df_h1_2 = pd.DataFrame() 
    h1 = 20
    
    
    for j in range (0,5):               # Смотрю куски по h1
        df_h1_1 = pd.DataFrame()
        df_h1_1 = df_Vx[(df_Vx['h1 [град]'] == h1)]
        

        df_n_1 = pd.DataFrame()
        df_n_2 = pd.DataFrame()
        n = 0

        for k in range (0,5):                                       # Смотрю куски по n1
            df_n_1 = df_h1_1[(df_h1_1['n1 [об/мин]'] == n)]
            n += 250
            df_n_2 = df_n_2.append(df_n_1, ignore_index=True)   
            
        df_h1_2 = df_h1_2.append(df_n_2, ignore_index=True)
        h1 += 2.5

    df_done = df_done.append(df_h1_2, ignore_index=True)    
    Vx += 2.5
# df_done.head(20)
```
