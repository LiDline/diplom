```python
import pandas as pd
import numpy as np
```

```python
file = 'C:/Users/Burger/source/work/double-propellers-and-rings/2СВК/T/Calculation Results/T1/3-я серия/0. 2 и 3-я серия (не сорт).csv'
new_df = pd.read_csv(file, sep = ',',skiprows=3)
new_df =new_df.apply(pd.to_numeric, errors = 'coerce')

df = pd.DataFrame(new_df.iloc[:,1:])
df.rename({'P28': 'Vн [км/ч]', 'P1': 'h1 [град]', 'P29': 'n1 [об/мин]', 'P2': 'h2 [град]', 'P30': 'n2 [об/мин]',
            'P17': 'ТВ1 [Н]','P18': 'ТВ2 [Н]', 'P19': 'RК1 [Н]','P20': 'RК2 [Н]',
             'P21': 'R ВМГ [Н]', 'P22': 'R тела [Н]','P23': 'R суммарная [Н]', 
             'P24': 'Mвx1 [Н*м]', 'P25': 'Mвx2 [Н*м]',  'P26': 'Mx ВМГ [Н*м]', 'P27': 'Mx суммарная [Н*м]',
             }, axis=1, inplace=True) 
df.loc[(df['n1 [об/мин]'] == 0.01), 'n1 [об/мин]'] = 0 
df.loc[(df['n2 [об/мин]'] == -0.01), 'n2 [об/мин]'] = 0    

df = df.reindex(columns=['Vн [км/ч]', 'h1 [град]', 'n1 [об/мин]', 'h2 [град]', 'n2 [об/мин]', 'ТВ1 [Н]','ТВ2 [Н]', 'RК1 [Н]','RК2 [Н]',
             'R ВМГ [Н]', 'R тела [Н]','R суммарная [Н]', 
            'Mвx1 [Н*м]', 'Mвx2 [Н*м]',  'Mx ВМГ [Н*м]', 'Mx суммарная [Н*м]',])

df.head()
```

```python
Vn_counter_0 = round((df['Vн [км/ч]'].max() - df['Vн [км/ч]'].min())/2.5 + 1)
h1_counter_0 = round((df['h1 [град]'].max() - df['h1 [град]'].min())/2.5 + 1)
n12_counter_0 = round((df['n1 [об/мин]'].max() - df['n1 [об/мин]'].min())/250 + 1)
```

```python
df_done = pd.DataFrame()

Vx_0 = df['Vн [км/ч]'].min()
Vx_step = 2.5

h1_step = 2.5

n_0 = df['n1 [об/мин]'].min()
n_step = 250

for i in range (0,Vn_counter_0):               # Смотрю куски по Vн
    df_Vx = pd.DataFrame()
    df_Vx = df[(df['Vн [км/ч]'] == Vx_0)]
    df_h1_2 = pd.DataFrame() 
    h1 = df['h1 [град]'].min()
    
    
    for j in range (0,h1_counter_0):               # Смотрю куски по h1
        df_h1_1 = pd.DataFrame()
        df_h1_1 = df_Vx[(df_Vx['h1 [град]'] == h1)]
        

        df_n_1 = pd.DataFrame()
        df_n_2 = pd.DataFrame()
        n = n_0

        for k in range (0,n12_counter_0 ):                                       # Смотрю куски по n1
            df_n_1 = df_h1_1[(df_h1_1['n1 [об/мин]'] == n)]
            n += n_step
            df_n_2 = df_n_2.append(df_n_1, ignore_index=True)   
            
        df_h1_2 = df_h1_2.append(df_n_2, ignore_index=True)
        h1 += h1_step

    df_done = df_done.append(df_h1_2, ignore_index=True)    
    Vx_0 += Vx_step
df_done.info(20)
```

```python
df_done.to_csv('Т1, Vн [0-27.5], серии 2 и 3.csv')
```
