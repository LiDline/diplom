### 2-ая таблица с данными

```python
file = 'C:/Users/Burger/source/work/double-propellers-and-rings/2СВК/T/Т1, Vн [0-27.5], серии 2 и 3.csv'
new_df = pd.read_csv(file, sep = ',')
new_df =new_df.apply(pd.to_numeric, errors = 'coerce')

df_2 = pd.DataFrame(new_df.iloc[:,1:])
df_1 = df_1.append(df_2)

df_1 = df_1.reindex(columns=['h1 [град]', 'h2 [град]', 'Vн [км/ч]', 'n1 [об/мин]', 'n2 [об/мин]', 'ТВ1 [Н]','ТВ2 [Н]', 'RК1 [Н]','RК2 [Н]',
             'R ВМГ [Н]', 'R тела [Н]','R суммарная [Н]', 
            'Mвx1 [Н*м]', 'Mвx2 [Н*м]',  'Mx ВМГ [Н*м]', 'Mx суммарная [Н*м]',])
# df_1 = df_1.iloc[:,:3]

df_1.loc[(df_1['n1 [об/мин]'] == 0), 'n1 [об/мин]'] = 0.01 
df_1.loc[(df_1['n2 [об/мин]'] == -0), 'n2 [об/мин]'] = -0.01  

df_1.head()
```

### Шаблон таблицы, сюда занесём известные данные


```python
file = 'C:/Users/Burger/source/work/double-propellers-and-rings/2СВК/for Ansys/Шаблон3.csv'
new_df = pd.read_csv(file, sep = ',',skiprows=3)
new_df.rename({'P28': 'Vн [км/ч]', 'P1': 'h1 [град]', 'P29': 'n1 [об/мин]', 'P2': 'h2 [град]', 'P30': 'n2 [об/мин]',}, axis=1, inplace=True) 
new_df = pd.DataFrame(new_df.iloc[:,1:]) 

h1 = 35
h2 = 35


new_df['h1 [град]'] = [float(h1)]*len(new_df['h1 [град]'])
new_df['h2 [град]'] = [float(-h2)]*len(new_df['h1 [град]'])

new_df = new_df.reindex(columns=['h1 [град]', 'h2 [град]', 'Vн [км/ч]', 'n1 [об/мин]', 'n2 [об/мин]'])

merge_df = pd.merge(new_df, df_1, how = 'left') 

merge_df.to_csv(f'. {h1}_{h2}, серия 4.csv')
```
